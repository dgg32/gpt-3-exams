```python
import os
import sys
import openai

question_file = "single_choice_english.txt"


openai.api_key = os.getenv("OPENAI_API_KEY")


def ask_question (text):
    
    instruction = "Choose the right word to fill the gap in the following sentence:\n"
    my_prompt = instruction + text

    print (my_prompt)

    response = openai.Completion.create(
        engine="text-davinci-002",
        prompt=my_prompt,
        temperature=0.7,
        max_tokens=256,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0
    )

    return response.choices[0].text.lstrip()

with open(question_file, 'r', encoding='utf-8') as file_in:
    content = ""
    for line in file_in:

        line = line.strip()
        content += line + "\n"
        if line.startswith("A"):
            
            res = ask_question(content)

            print(f"GPT-3: {res}\n\n")

            content = ""


```
